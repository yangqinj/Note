# Spark架构

spark的架构主要有三部分组成：`SparkContext`（驱动程序）、`ClusterManager`（集群资源管理器）和`Executor`（任务执行进程）。

Screen Shot 2019-09-13 at 14.05.54.png

+ SparkContext提交作业，向ClusterManager申请Executor资源
+ ClusterManager根据当前集群的资源使用情况，使用FIFO策略进行资源分配
+ ClusterManager会尽可能将程序分布在尽可能多的Worker上，从而有利于充分利用集群资源，适合内存使用多的场景；另一种是将程序分布在尽可能少的Worker上，适合CPU密集型场景。
+ 分配资源后，SparkContext和Executor保持通讯，任务分发给Executro，由Executor执行任务。


SparkContet负责整个任务的管理，包括：连接Spark集群、创建RDD、累加器、广播变量等

## transforms & actions

* **`map()` 对RDD的每一个item都执行同一个操作**
* **`flatMap()` 对RDD中的item执行同一个操作以后得到一个list，然后以平铺的方式把这些list里所有的结果组成新的list**
* **`filter()` 筛选出来满足条件的item**
* **`distinct()` 对RDD中的item去重**
* **`sample()` 从RDD中的item中采样一部分出来，有放回或者无放回**
* **`sortBy()` 对RDD中的item进行排序**

````python
numbersRDD = sc.parallelize(range(1, 11))
print(numbersRDD.collect())

squareRDD = numbersRDD.map(lambda x: x * x)
print(squareRDD.collect())
````

如果你手头上有2个RDD了，下面的这些操作能够帮你对他们以个种方式组合得到1个RDD:
* **`rdd1.union(rdd2)`: 所有`rdd1`和`rdd2`中的item组合**
* **`rdd1.intersection(rdd2)`: `rdd1` 和 `rdd2`的交集**
* **`rdd1.substract(rdd2)`: 所有在`rdd1`中但不在`rdd2`中的item（差集）**
* **`rdd1.cartesian(rdd2)`: `rdd1` 和 `rdd2`中所有的元素笛卡尔乘积 **

````python
numbersRDD.union(squareRDD).collect()
````

````python
numbersRDD.intersection(squareRDD).collect()
````

**特别注意**：Spark的一个核心概念是**惰性计算**。当你把一个RDD转换成另一个的时候，这个转换不会立即生效执行！！！<br>Spark会把它先记在心里，等到真的需要拿到转换结果的时候，才会重新组织你的transformations(因为可能有一连串的变换)<br>
这样可以避免不必要的中间结果存储和通信。

刚才提到了**惰性计算**，那么什么东西能让它真的执行转换与运算呢？
是的，就是我们马上提到的**Actions**，下面是常见的action，当他们出现的时候，表明我们需要执行刚才定义的transform了:
* `collect()`: 计算所有的items并返回所有的结果到driver端，接着 `collect()`会以Python list的形式返回结果
* `first()`: 和上面是类似的，不过只返回第1个item
* `take(n)`: 类似，但是返回n个item
* `count()`: 计算RDD中item的个数
* `top(n)`: 返回头n个items，按照自然结果排序
* `reduce()`: 对RDD中的items做聚合

````python
numbersRDD.reduce(lambda x, y: x+ y)
````

````python
numbersRDD.reduce(lambda x, y: x * y)
````

````python
file_rdd.filter(lambda x: x[0] == 'M').map(lambda x: x.split(',')[0]).reduce(lambda x, y: x + ', ' + y)
````

````python
# 使用cache暂存中间结果
squareRDD.cache()
````

````python
# 计算平均值
# 这样只是【激活】了两次squareRDD，而不是【计算】了两次squareRDD
# 因为reduce和count是两个action
squareRDD.reduce(lambda x, y: x + y) / squareRDD.count()
````

## 更复杂的transformations和actions

有时候我们会遇到更复杂的结构，比如非常非常经典的是以元组形式组织的k-v对（key, value）<br>
我们把它叫做**pair RDDs**，而Sark中针对这种item结构的数据，定义了一些transformation和action:
* `reduceByKey()`: 对所有有着相同key的items执行reduce操作
* `groupByKey()`: 返回类似(key, listOfValues)元组的RDD，后面的value List 是同一个key下面的
* `sortByKey()`: 按照key排序
* `countByKey()`: 按照key去对item个数进行统计
* `collectAsMap()`: 和collect有些类似，但是返回的是k-v的字典

````python
str_rdd = sc.parallelize(["Hello hello", "Hello New York", "York says hello"])

word_cnt = (str_rdd.flatMap(lambda sent: sent.split())
              .map(lambda word: word.lower())
              .map(lambda word: (word, 1))
              .reduceByKey(lambda x, y: x + y))
````

````python
# 以字典形式返回结果
word_cnt.collectAsMap()
````

````python
# 出现频率最高的两个词
word_cnt.sortBy(keyfunc=lambda wcnt: wcnt[1], ascending=False).take(2)
````

## 编程的方式

````python
from pyspark.sql.types import *
````

````python
sc = spark.sparkContext
````

````python
# 文件最后不能有空行！！！否则会解析错误
lines = sc.textFile('../datasets/people.txt') # lines是一个RDD
parts = lines.map(lambda l: l.split(","))
# 数据转换为tuple
people = parts.map(lambda p: (p[0], int(p[1]))) 
````

````python
people
````

````python
# 构建schema
schema_cols = 'name age'
fileds = [StructField(name, StringType(), True) for name in schema_cols.split()]
schema = StructType(fileds)
````

````python
# 使用schema将RDD转化为DataFrame
people_df = spark.createDataFrame(people, schema)
````

````python
people_df
````

````python
# 将DataFrame转化为Dataset
people_df.createOrReplaceTempView('people')
````

````python
spark.sql('SELECT * FROM people').show()
````

+ RDD是Spark中操作的基础数据单位，存储的是一个一个的数据。数据可以是结构化的，也可能是非结构化的。
+ DataFrame则是在RDD的基础上进行的封装，其多了schema结构，表明了数据的结果信息，从而提升运行效果。DataFrame中存储的是结构化数据，类似SQL数据集。
+ Dataset则是DataFrame的抽象，在编译期间检查数据类型，并且使用Encoder编码数据和传输数据。
+ 现在，DataFrame和Dataset的api接口是统一的。
