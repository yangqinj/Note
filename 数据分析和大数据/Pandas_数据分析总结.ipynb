# 时间信息处理

## 读取数据格式

```python
bikes = pd.read_csv('../datasets/bike.csv', usecols=usecols,
                   parse_dates=['datetime'], dtype=dtypes)
```

## 获取年、月、小时、星期几信息

```python
bikes['year'] = bikes.datetime.dt.year
```

```python
bikes['month'] = bikes.datetime.dt.month
```

```python
bikes['hour'] = bikes.datetime.dt.hour
```

```python
import calendar
```

```python
bikes['weekday'] = bikes.datetime.apply(lambda x : calendar.day_name[x.weekday()])
```

## 时间数据可视化：单因素

当特征中有时间时，可以观察目标变量随着时间的变化，包括：折线图、柱状图、箱线图等。假设考虑不同月份租车人数的变化：

```python
fig = plt.figure(figsize = (18, 5))

# 柱状图
ax_barplot = fig.add_subplot(131)
sns.barplot(x='month', y='count', data=bikes, palette='deep')
ax_barplot.set(ylabel = 'count', title = 'Count over Month')

# 折线图
ax_pointplot = fig.add_subplot(132)
sns.pointplot(x='month', y='count', data=bikes)
ax_pointplot.set(ylabel = 'count', title = 'Count over Month')

# 箱线图
ax_boxplot = fig.add_subplot(133)
sns.boxplot(x='month', y='count', data=bikes)
ax_boxplot.set(ylabel = 'count', title = 'Count over Month')
```

# 多因素分析

当需要考虑多因素对目标变量的影响时，需要根据因素的类型（数值型、分类型），特征不同值的个数来决定需要使用哪种。可以考虑：分组柱状图、堆积柱状图（观察不同因素的占比）、分组折线图、分组箱线图等。

## 分组计数

```python
bikes.groupby(['season', 'weather'])['season'].count().unstack(['weather'])
```

## 不同年份和季节对日期的影响： 2对1

数据集中只有两个年份（2011和2012）和四个季节，都是分类型数据。可以考虑使用分组柱状图、分组折线图和分组箱线图

```python
bikes['ones'] = 1
```

```python
fig = plt.figure(figsize = (18, 5))
fig.suptitle('ShareBike Analysis', fontsize=14, fontweight='bold')

# 柱状图
ax_barplot = fig.add_subplot(131)
sns.barplot(x='season', y='count', hue='year', data=bikes, palette='deep')
ax_barplot.set(ylabel = 'count', title = 'Count over Season and Year')


# 折线图
ax_pointplot = fig.add_subplot(132)
sns.pointplot(x='season', y='count', hue='year', data=bikes)
ax_pointplot.set(ylabel = 'count', title = 'Count over Season and Year')


# 箱线图
ax_boxplot = fig.add_subplot(133)
sns.boxplot(x='season', y='count', hue='year', data=bikes)
ax_boxplot.set(ylabel = 'count', title = 'Count over Season and Year')
```

# 连续特征离散化

经常需要对连续性的特征进行分析，但是连续变量需要进行离散化。离散化分为等频离散化和等宽离散化。等频离散化将特征等分，每个桶中数据的个数一样。等宽离散化将特征等分，每个桶中数据的宽度（上下界间隔一样）。例如对温度一列进行等频离散化

## 等频离散化

```python
bikes.temp.describe()
```

```python
bikes['cut_freq'] = pd.cut(bikes.temp.values, bins=5)
```

```python
bikes.head()
```

```python
sns.barplot('cut_freq', 'count', data=bikes)
```

保证每个桶内的数据都是一样的，但是会将相同的值放入到不同的桶中

## 等宽离散化

```python
bikes['cut_width'] = pd.cut(bikes.temp, 5)
```

```python
bikes
```

可以看到，温度一列大概是每8度一个桶。

```python
sns.barplot('cut_width', 'count', data=bikes)
```

## 按照给定的点分割数据

```python
bikes['cut_point'] = pd.cut(bikes.temp, bins=[0.78, 10, 100])
```
